```python
from scipy import stats
import matplotlib.pyplot as plt

fig = plt.figure()
ax1 = fig.add_subplot(211)

X = stats.loggamma.rvs(5, size=500)+5
prob = stats.probplot(X, dist= 'norm', plot=ax1)
ax1.set_xlabel('')
ax1.set_title("Probplot against normal distribution")

ax2 = fig.add_subplot(212)
Xt,_= stats.boxcox(X)
prob = stats.probplot(Xt, dist= 'norm', plot=ax2)
ax2.set_title("Probplot after BoxCox transformation")

plt.show()
```

```python

import numpy
from pandas import read_csv
from sklearn.utils import resample
from matplotlib import pyplot

X = numpy.array([180,162,172,158,168,150,171,183,165,176])

n_iterations = 1000
n_size = int(len(X))


medians = list()

for i in range(n_iterations):
    S= resample(X,n_samples=n_size)
    m = numpy.median(S)
    
    medians.append(m)

pyplot.hist(medians)
pyplot.show()

#confidence intervals
 
a = 0.95
p = ((1-a)/2.0)*100
lower = numpy.percentile(medians,p)

p=(a+((1-a)/2.0))*100
upper = numpy.percentile(medians,p)

print('%.1f Confidence interval %.1f and %.1f' % (a*100,lower, upper))
```

```python
## This is test X is a normal distribution or not

import numpy as np
import seaborn as sns
from scipy import stats
import matplotlib.pyplot as plt

X = stats.norm.rvs(size = 1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(X),bw=0.5)
plt.show()
```

```python
stats.kstest(X, 'norm')

# P-Value is 73% so H0 is accepted hence X is a gaussian distribution
```

```python
Y = np.random.uniform(0,1,10000)
sns.kdeplot(np.array(Y),bw=0.1)
plt.show()
```

```python
stats.kstest(Y, 'norm')
# P-Value is 0% so H0 is rejected hence X is not a gaussian distribution. Both X and Y has the different distributions.
```
